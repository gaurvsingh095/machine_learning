```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
area_df = df.drop('price',axis='columns')
area_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(area_df,price)
```

```python
reg.predict([[3300]]) #(1) Predict price of a home with area = 3300 sqr ft
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#Y = m * X + b (m is coefficient and b is intercept)
3300*135.78767123 + 180616.43835616432
```

```python
reg.predict([[5000]]) #(1) Predict price of a home with area = 5000 sqr ft
```

```python
#Generate CSV file with list of home price predictions
predict_area_df = pd.read_csv("areas.csv")
predict_area_df.head(3)
```

```python
p = reg.predict(predict_area_df)
p
```

```python
predict_area_df['prices']=p
predict_area_df
```

```python
predict_area_df.to_csv("prediction.csv")
```
